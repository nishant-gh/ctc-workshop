# Exercise 1: Your First API Call

In this exercise, you'll make your first call to the Gemini API using the Google GenAI Python SDK.

By the end, you'll understand:
- How to create a GenAI client
- The basic structure of a `generate_content` call
- What the API response looks like

## Step 1: Create the client

The setup cell above loads your API key from Colab Secrets.


```python
from google import genai

client = genai.Client()
```

## Step 2: Send a message

Every API call needs at minimum:
- `model` — which model to use
- `contents` — the prompt text or a list of `Content` objects

```python
response = client.models.generate_content(
    model="gemini-3-flash-preview",
    contents="What are three fun facts about the moon?",
)
```

## Step 3: Explore the response

Let's look at the full response object first to understand its structure.

```python
response
```

The response contains:
- `candidates` — a list of completions (usually one), each with `content`
- `usage_metadata` — token counts for input and output

Each candidate's content has:
- `parts` — a list of parts (text, function calls, etc.)
- `role` — always `'model'` for assistant responses

To get just the text:

```python
print(response.text)
```
